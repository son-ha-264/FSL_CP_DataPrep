```python
import sqlite3
import pandas as pd
```

```python
# Init path
chembl_db_path = "/home/son.ha/FSL_CP_DataPrep/sql/chembl_31_sqlite/chembl_31.db"
```

```python
# Query
query_assay_target = "SELECT DISTINCT ay.assay_id, ay.chembl_id, t.chembl_id, t.target_type " \
                                "from assays ay, target_dictionary t " \
                                "where ay.tid = t.tid "
```

```python
# Fetch 
db = sqlite3.connect(chembl_db_path)
cursor = db.cursor()
myresult = cursor.execute(query_assay_target).fetchall()


df_target = pd.DataFrame(myresult, columns=['ASSAY_ID', 'assay_chembl_id', 'target_chembl_id', 'target_type'])
df_target
```

```python
df_cp_assay = pd.read_csv("temp/chembl31-labels-assay-index.csv")
```

```python
df_cp_assay
```

```python
df_merge = pd.merge(df_cp_assay, df_target, on="ASSAY_ID", how="left")
df_merge
```

```python
df_assay_2 = pd.read_csv("temp/chembl31-labels-assay-index-nondup.csv")
```

```python
df_merge["ASSAY_ID"] = df_assay_2["ASSAY_ID"]
```
